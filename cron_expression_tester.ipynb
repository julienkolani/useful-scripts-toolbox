```python
from croniter import croniter
from datetime import datetime, timedelta

def compare_intervals(cron_expression, test_date=None):
    if test_date is None:
        test_date = datetime.now()

    # Créer un objet croniter avec l'expression cron et la date de test
    cron = croniter(cron_expression, test_date)
    print(f"Expression cron: {cron_expression}")

    # Obtenir la future date du cron
    future_execution_date = cron.get_next(datetime)
    print(f"Future execution date: {future_execution_date}")

    # Obtenir l'ancienne date du cron
    old_execution_date = cron.get_prev(datetime)
    print(f"Old execution date: {old_execution_date}")

    # Obtenir la date précédente du précédent
    previous_previous_execution_date = cron.get_prev(datetime)
    print(f"Previous previous execution date: {previous_previous_execution_date}")

    # Calculer les intervalles
    interval_between_future_and_old = future_execution_date - old_execution_date
    interval_between_previous_and_previous_previous = old_execution_date - previous_previous_execution_date
    print(f"Interval between future and old: {interval_between_future_and_old}")
    print(f"Interval between previous and previous previous: {interval_between_previous_and_previous_previous}")
    
    # Convertir le seuil de similarité en timedelta
    similarity_threshold = timedelta(minutes=1)
    print(f"Similarity threshold: {similarity_threshold}")

    # Comparer les intervalles
    intervals_are_similar = abs(interval_between_future_and_old - interval_between_previous_and_previous_previous) < similarity_threshold
    print(f"Intervals are similar: {intervals_are_similar}")

    return not intervals_are_similar

def run_cron_comparison_tests():
    # Liste d'expressions cron à tester
    cron_expressions = [
        "00 12 * * 5"
    ]

    # Variables pour la date de test
    test_year = 2024
    test_month = 1
    test_day = 12
    test_hour = 12
    test_minute = 0
    test_second = 0

    test_date = datetime(test_year, test_month, test_day, test_hour, test_minute, test_second)

    for cron_expression in cron_expressions:
        print(f"Testing cron expression: {cron_expression}")
        result = compare_intervals(cron_expression, test_date)
        print(f"Result: {result}")
        print("-" * 30)

if __name__ == "__main__":
    run_cron_comparison_tests()
    
```
